```python
import ultralytics
import cv2
import numpy as np
import pandas as pd
import time
import pickle
```

```python
with open('cached_model.pkl','rb') as f:
    model = pickle.load(f)
```

```python
cap = cv2.VideoCapture('highway_mini.mp4')
while True:
    ret,frame = cap.read()
    if not ret:
        print("breaking the loopp")
        break

    frame = cv2.resize(frame,(1020,500))
    results = model(frame)
    for box in results[0].boxes.data:
        box = box.detach().cpu().numpy()
        x,y,w,h,d,_ = box
        print("x:", x, "y:", y, "w:", w, "h:", h) 
        x2 = int(x+w)
        y2 = int(y+h)
        cv2.rectangle(frame,(int(x),int(y)),(x2,y2),(0,255,0),2)
    cv2.imshow('frame',frame)
        
    # if cv2.waitKey(10) & 0xFF == 27:
    if cv2.waitKey(0) & 0xFF == 27:
        break
cap.release()

cv2.destroyAllWindows()
    
```
